# Zmienne

```python
# liczba całkowita 'int'
calkowita = 10

# ciąg znaków 'str'
tekst = "Jakiś tekst"

# liczba zmiennoprzecinkowa 'float'
zmiennoprzecinkowa = 10.5

# zmienna logiczna 'bool'
logiczna = True

print(f"Zwykły tekst {calkowita} {tekst} {zmiennoprzecinkowa} {logiczna}")
print(f"Zmienna calkowita jest typu {type(calkowita)}")
```

```python
lista = [calkowita, tekst, zmiennoprzecinkowa, logiczna]

print(lista)
```

# Instrukcje warunkowe

```python
LOGIN = "gigant"
HASLO = "1234"

login = input("Podaj login: ")
if login == LOGIN:
  haslo = input("Podaj hasło: ")
  if haslo == HASLO:
    print("Zalogowano!")
  else:
    print("Złe hasło!")

else:
  print("Zły login!")
```
